```python
# PASSWORD
import sys
from PyQt5.QtWidgets import QMainWindow, QApplication, QWidget, QPushButton, QAction, QLineEdit, QMessageBox, QLabel
from PyQt5.QtGui import QIcon, QFont
from PyQt5.QtCore import pyqtSlot
from PyQt5 import QtCore, QtGui
import random

class App(QMainWindow):
  def __init__(self):
    super().__init__()
    self.title = '컴투존피씨방에서 만원에 구입한 싸구려 보안 프로그램'
    self.left = 0
    self.top = 0
    self.width = 1920
    self.height = 1080
    self.initUI()
 
  def initUI(self):
    self.setWindowTitle(self.title)
    self.setGeometry(self.left, self.top, self.width, self.height)
 
    self.label = QLabel('님컴퓨터이제못함ㅋ\n비밀번호입력하셈 -엄마', self)
    self.label.setFont(QFont("Arial", 20))
    self.label.resize(280, 200)
    self.label.move(640, 160)

    # Create textbox
    self.textbox = QLineEdit(self)
    self.textbox.setEchoMode(QLineEdit.Password)
    self.textbox.setFont(QFont("Arial", 20))
    # self.textbox.setAlignment(QtCore.Qt.AlignCenter)
    self.textbox.move(580, 300)
    self.textbox.resize(280, 40)
 
    # Create a button in the window
    self.button = QPushButton('확인', self)
    self.button.move(670, 360)
 
    # connect button to function on_click
    self.button.clicked.connect(self.on_click)
    self.textbox.returnPressed.connect(self.on_click)
    self.show()
 
  @pyqtSlot()
  def on_click(self):
    textboxValue = self.textbox.text()
    if len(textboxValue) < 2 or len(textboxValue) > 10:
      QMessageBox.question(self, '엄마로부터의 메시지', '비밀번호는 최소 2자리, 최대 10자리입니다', QMessageBox.Ok, QMessageBox.Ok)
    else:
      rand = random.randint(0, 100)
      QMessageBox.question(self, '', '비밀번호 일치 점수: %s' % (rand), QMessageBox.Ok, QMessageBox.Ok)
    self.textbox.setText('')

if __name__ == '__main__':
  app = QApplication(sys.argv)
  ex = App()
  sys.exit(app.exec_())
```

```python
## Genetic hacking password

import random
import string


```

```python
password = 'brad1234'
min_len = 2
max_len = 10


def generate_word(len=''):
    return ''.join(random.sample(string.ascii_letters + string.digits, k=len))

def generate_population(size, min_len, max_len):
    population = []

    for i in range(size):

        len = i % (max_len - min_len + 1) + min_len
        population.append(generate_word(len))

    return population

# print(generate_word(10))
pop = generate_population(100, min_len, max_len)
print(pop)
```

```python
# Assesment password length and characters
# password: Answer
# test_word: from population
def compute_marks(password, test_word):
    mark = 0

    if len(password) != len(test_word):
        return mark

    len_mark = 0.5
    mark += len_mark

    for idx, val in enumerate(list(str(password))):
        if val == test_word[idx]:
            mark += 1

    return mark / (len(password) + len_mark) * 100

compute_marks('ab23', 'ab23') 
```

```python
# Compute performance, alive children

def compute_performance(population, password):
    performance_list = []
    len_password = 0

    for child in population:
        mark = compute_marks(password, child)

        if mark > 0:
            len_password = len(child)

        performance_list.append([child, mark])

    # RANKING 
    population_sorted = sorted(performance_list, key=lambda x : x[1], reverse=True)

    return population_sorted, len_password


def select_super(population_sorted, best_sample, lucky_few, len_password):
    next_gen = []

    for i in range(best_sample):
        
        # selection best performance for next generation 
        if population_sorted[i][1] > 0:
            next_gen.append(population_sorted[i][0])

        lucky_population = random.sample(population_sorted, k=lucky_few)
        for ll in lucky_population:
            next_gen.append(ll[0])


        # if not enough next generation, randomly generate
        while len(next_gen) < best_sample + lucky_few:
            next_gen.append(generate_word(len_password))


    return next_gen 


population_sorted, len_password = compute_performance(pop, password)
# print(population_sorted) 
super_pop = select_super(population_sorted, best_sample=20, lucky_few=20, len_password=len_password)
# print(super_pop)
                            
        
```

```python
# Create Child
def create_child(child1, child2):
    child = ''

    # in case of being different 
    min_len_child = min(len(child1), len(child2))

    for i in range(min_len_child):

        # 50 percent child1
        if (int(100* random.random()) < 50):
            child += child1[i]
        else:
            child += child2[i]

    return child

def create_children(parents, n_child):
    next_pop = []

    # Only half couple would be able to get married 
    for i in range(int(len(parents) / 2)):
        for j in range(n_child):
            next_pop.append(create_child(parents[i], parents[len(parents) - 1 - i]))

    return next_pop

children = create_children(super_pop, 5)

print(children)
                   
```

```python
# Mutation
# randomly located word to random word 
def mutation_word(word):
    idx = int(random.random() * len(word))

    if (idx == 0):
        word = random.choice(string.ascii_letters + string.digits) + word[1:]

    else:
        word = word[:idx] + random.choice(string.ascii_letters + string.digits) + word[idx+1:]

    return word

def mutation_population(population, chance_of_mutation):
    for i in range(len(population)):
        if random.random() * 100 < chance_of_mutation:
            population[i] = mutation_word(population[i])

    return population

next_generation = mutation_population(population=children, chance_of_mutation=10)

print(next_generation)
```

```python
# test

password = 'tt12334'
n_generation = 300
population = 100
best_sample = 20
lucky_few = 20
n_child = 5
chance_of_mutation = 10

pop = generate_population(population, min_len, max_len)

for g in range(n_generation):
    pop_sorted, len_password = compute_performance(population=pop, password=password)

    if int(pop_sorted[0][1]) == 100:
        print('Success the password is : ', pop_sorted[0][0])
        break

    super_pop = select_super(pop_sorted, best_sample, lucky_few, len_password)
    children = create_children(super_pop, n_child)
    new_generation = mutation_population(children, 10)

    pop = new_generation

    print('======= %sth Generation========' %(g + 1))
    print(pop_sorted[0])
```
